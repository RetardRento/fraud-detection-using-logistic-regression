```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
card_data = pd.read_csv('creditcard.csv')
card_data.head()
```

```python
card_data.tail()
```

```python
card_data.info()
```

```python
card_data.isna().sum()
```

```python
card_data['Class'].value_counts()
```

```python
legit = card_data[card_data.Class == 0]
fraud = card_data[card_data.Class == 1]
```

```python
print(legit.shape , fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
card_data.groupby('Class').mean()
```

```python
#Undersampling the dataset
#create a test data set with equal amounts of normal and fruad transactions
```

```python
legit_sample = legit.sample(n = 492)
```

```python
new_dataset = pd.concat([legit_sample , fraud] , axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset
```

```python
#Splitting the data into dependent and Independent Sets
#Basically into Features and Targets
```

```python
x = new_dataset.drop(columns = 'Class' , axis = 1)
y = new_dataset['Class']
```

```python
x
```

```python
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , stratify = y , random_state = 2)
```

```python
print(x.shape , x_train.shape , x_test.shape)
```

```python
# Model Training
# Logistic Regression
```

```python
model = LogisticRegression()
```

```python
#training the model with training data_set
model.fit(x_train , y_train)
```

```python
# Evaluvation
```

```python
#accuracy on training data

x_train_predict = model.predict(x_train)
train_accuracy = accuracy_score(x_train_predict , y_train)
```

```python
print('Training data accuracy: ' , train_accuracy)
```

```python
y_test.shape
```

```python
#accuracy on test data
x_test_predict = model.predict(x_test)
x_test_accuracy = accuracy_score(x_test_predict , y_test)
```

```python
print('Testing data accuracy: ' , x_test_accuracy)
```
